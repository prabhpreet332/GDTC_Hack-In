```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
```

### Click a pic

```python
capture = cv2.VideoCapture(0)
while True:
    vid = capture.read()
    im_path = 'captured_images/img.jpg'
    cv2.imwrite(im_path, vid[1])
    print("Image Clicked!")
    cv2.waitKey(1)
    break
capture.release()
cv2.destroyAllWindows()
```

```python
import os
image = cv2.imread(os.getcwd() + '/captured_images/img.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### Continuously click pictures (clicks specified number of pictures)


```python
cap = cv2.VideoCapture(0)
num = 5 # This is the number of Images to be clicked

while True:
    ret,frame = cap.read()

    for count in range(num):
        im_path = f'captured_images/img_cont_{count}.jpg'
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        cv2.imwrite(im_path, frame)
        time.sleep(1)
        print(f"Image {count+1} captured")
        if count == num-1:
            break

    cv2.waitKey(1)
    break

cap.release()
cv2.destroyAllWindows()
```
