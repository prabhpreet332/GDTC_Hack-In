```python
import re
import pandas as pd
import PyPDF2
import csv
```

```python
#Extract text from the pdf first
pdfFileObj = open('second_sem.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)  
print(pdfReader.numPages) 
```

```python
pageData = []
for i in range(0,277):
    pageObj = pdfReader.getPage(i)
    pageData.append(pageObj.extractText())
```

```python
pageDataCleaned = []
for pg in pageData:
    pageDataCleaned.append(re.findall(r'\d+\n\n.*\n\nSID: \d+\n\nSchemeID: \d+\n99102.4.\n.*\n.*\n99104.3.\n.*\n.*\n28106.3.\n.*\n.*\n27108.3.\n.*\n.*\n36110.3.\n.*\n.*\n98112.3.\n.*\n.*\n56114.3.\n.*\n.*\n99152.1.\n.*\n.*\n27154.1.\n.*\n.*\n28156.1.\n.*\n.*\n36158.1.\n.*\n.*\n56160.1.\n.*\n.*\n.*\n',pg))
```

```python
data_raw = []
#Replacing the '\n' character being visible due to usage of raw string in regular expression
for pg in pageDataCleaned:
    for p in pg:
        newP = p.replace('\n',',')
        data_raw.append(newP)
```

```python
student = data_raw[0].split(',')
z = student[1]
student = [m for m in student if m!=z]
len(data_raw)
```

```python
final_student_data = []
for dr in data_raw:
    dummy_student = dr.split(",")
    dummy_student = [ds for ds in dummy_student if ds != ""] #Helps remove all of the empty elements of the list
    final_student_data.append(dummy_student)
#To sort out this data more we are going to divide this data into subjects along as internal and external marks
#We directly know what indexes we have to fix
index_subjects = [5,8,11,14,17,20,23,26,29,32,35,38]
#Presence of college leavers create a hole in the indentation so trading off vipul's data with muskan's
len(final_student_data)
```

```python
absents = []
for fsd in final_student_data:
    for f in fsd:
        if f in ['A99104(3)','A27108(3)','AA','A 0 ','0 A ','A98112(3)','A28106(3)']:
            print(fsd)
            final_student_data.remove(fsd)
            break
```

```python
for fsd in final_student_data:
    print(fsd)
print(len(final_student_data))
```

```python
marks_cleaned_students = []
for fsd in final_student_data:
    mcs = fsd[:4]
    for i in range(4,41):
        if i in index_subjects:
            try:
                temp_marks = fsd[i].split(" ")
                temp_marks = [t for t in temp_marks if t!= ""]
                mcs.append(temp_marks[0])
                mcs.append(temp_marks[1])
            except:
                print(fsd)
                print("Exceptions")
        else:
            mcs.append(fsd[i])
    marks_cleaned_students.append(mcs)
```
